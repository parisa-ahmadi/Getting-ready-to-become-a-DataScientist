```python
plt.figure(figsize=(9,7),dpi=80)
plt.plot(years,iran_pop)
plt.xlabel("years",size=20)
plt.ylabel("iran\'s popuation",size=20)
plt.xticks([1960,1970,1980,1990,2000,2010,2016])
plt.yticks([21.91,28.51,38.23,56.23,66.13,74.57,80.28],['21M','28M','38M','56M','66M','74M','80M'])
plt.show()
```

```python
city_name=['Tehran','Mashhad','Isfahan','Karaj','Tabriz','Shiraz','Bandare-Anzali']

city_pop=[7153309,2307177,1547164,1448075,1424641,1249942,1002316]

plt.hist(city_pop)
```

```python
pop_size=np.array([5153309, 2307177, 1547164, 1448075, 1424641, 1249942, 1002316])/2500
colors=["black","brown","blue","yellow","red","pink","purple"]
plt.scatter(np.arange(7),city_pop,s=pop_size,c=colors)
plt.margins(0.3)
plt.title("Iran\'s most populated city",size=15)
plt.xlabel("city\'s name",size=20)
plt.ylabel("city\'s population", size=20)
plt.xticks([0,1,2,3,4,5,6],['Tehran','Mashhad','Isfahan','Karaj','Tabriz','Shiraz','Bandare-Anzali'])
plt.yticks([1000000,2000000,3000000],["1M","2M","3M"])
plt.text(0,5153309,"Iran_capital",size=15)
plt.text(6,1002316,"Abay",size=15)
plt.show()
```

```python
turkey_pop=[27.47,34.88,43.98,53.92,63.24,72.33,79.51]

plt.plot(years,iran_pop, ls='-', marker='+' ,mew=8)

plt.plot(years,turkey_pop, ls='--', lw=1)#??????

plt.title('Iran VS Turkey Population', size=15 )

plt.legend(['Iran','Turkey'], loc='best')

plt.xlabel('Years', size=20)

plt.ylabel('Population', size=20)

plt.xticks([1960,1970,1980,1990,2000,2010,2016])

plt.yticks([20,30,40,50,60,70,80],['20M','30M','40M','50M','60M','70M','80M'])

plt.grid()

plt.annotate('Iran-Iraq War', xytext=(1990,34.88),fontsize=15,xy=(1980,38.67),arrowprops=dict(facecolor='silver',width=4))#????

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(years,iran_pop)
plt.title("Iran\'s population")
plt.subplot(1,2,2)
plt.plot(years,turkey_pop)
plt.grid()
plt.title("Turkey population")
plt.plot(turkey_pop,color='green')
plt.show()
```
