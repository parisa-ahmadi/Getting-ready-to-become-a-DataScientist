```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

record=pd.read_csv('data.csv')
record=record.sort_values(by='Calories')

np_arr=record.values
x_2=np_arr[:,0]
y_2=np_arr[:,1]
fig_4=plt.figure(figsize=(6,4))
axes_4=fig_4.add_axes([0,0,1,1])

axes_4.set_xlabel('Units Sold')
axes_4.set_ylabel('100 record')
axes_4.set_title('100 sales record')

axes_4.plot(x_2,y_2)

plt.show()
```

```python
ice_df=pd.read_csv('data.csv')
ice_df=ice_df.sort_values(by='Duration')

np_arr=ice_df.values
x_2=np_arr[:,0]
y_2=np_arr[:,1]
fig_4=plt.figure(figsize=(6,4))
axes_4=fig_4.add_axes([0,0,1,1])

axes_4.set_xlabel('Temperature')
axes_4.set_ylabel('Data')
axes_4.set_title('calories')

axes_4.plot(x_2,y_2)

axes_4.annotate('Best Month', xy=(83,500),xytext=(60,500),
               arrowprops=dict(facecolor='black',shrink=0.08))
```
