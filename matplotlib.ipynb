```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,5,10)
y=x**2
plt.plot(x,y)
```

```python


x=np.linspace(0,5,10)
y=x**2
plt.plot(x,y)
plt.title("Days Chart")
```

```python


x=np.linspace(0,5,10)
y=x**2
plt.plot(x,y)
plt.xlabel("Days")
plt.ylabel("Days squared")
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")
plt.subplot(1,2,2)
plt.plot(x,y,"b")
```

```python
fig_1=plt.figure(figsize=(5,4),dpi=100)
axes_1=fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Days squared')
axes_1.set_title('Days chart')
axes_1.plot(x,y,label="x**/x")
axes_1.plot(y,x,label="x/x**2")
axes_1.legend(loc="best")
plt.show()
```
