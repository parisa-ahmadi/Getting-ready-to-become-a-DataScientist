```python
#Code By Parisan.Ahmadi
import pandas as pd
import seaborn as sb
df=pd.read_csv("data.csv")
df
```

```python
sb.lmplot(x="Calories" , y="Maxpulse", data=df)
```

```python
#simple feature scaling(Normalization)
df["Calories"]=df["Calories"]/df["Calories"].max()
df
```

```python
#Xnew=Xold/Xmax
df["Maxpulse"]=df["Maxpulse"]/df["Maxpulse"].max()
df
```

```python
#Max-Min(Normalization)
#Xnew=Xold - Xmin / Xmax - Xmin
df["Calories"]=(df["Calories"]-df["Calories"].min())/(df["Calories"].max()-df["Calories"].min())
df
```

```python
#Z-score(Normalization)
#Xnew=Xold - M(Meo or mean) / Standard Division(Std)
df["Calories"]=(df["Calories"] - df["Calories"].mean())/df["Calories"].std()
print(df.to_string())
```
